### Wave Dark Matter (WDM) dynamics: Schrodinger-Poisson Equation


\begin{align}
i {\partial \psi\over \partial t} 
&= - {1 \over 2 \mu} \nabla^2\psi + g |\psi|^2 \psi + \mu \psi \Phi ~,
\\
\nabla^2 \Phi &= 4 \pi G \mu \left( |\psi|^2 - n_0 \right) ~,
\end{align}
where
- $g={ \lambda \over 8 \mu^2 }$: dimensional ([mass]$^{-2}$) coupling constant; $V(\phi) = {1 \over 2} \mu^2\phi^2 + { \lambda \over 4!} \phi^4$;
- $\Phi$: the gravitational potential;
- $\nabla^2 \equiv \delta^{ij} \partial_i \partial_j$: the spatial Laplace operator.

**Our task is to solve the wave function $\psi(t,\vec{x})$ from SP eq. with the given initial condition $\psi(t_0,\vec{x})$ and boundary condition (open boundary).**


### Dimensionless \& Units

The scaling symmetry of SP eq. is given by
\begin{align}
\big\{ t, \mathbf{r}, \psi, \Phi, g \big\} 
\rightarrow
\big\{ \lambda^{-2} t, \lambda^{-1} \mathbf{r}, \lambda^{2} \psi, \lambda^{2} \Phi, \lambda^{-2} g \big\} ~.
\end{align}


Introducing
\begin{align}
\mathcal{L} &\equiv \left( 8\pi \hbar^2 \over 3 m^2 H_0^2 \Omega_{m0} \right)^{1/4}
\simeq 121 \text{kpc} \left( 10^{-23} \text{eV} \over m \right)^{1/2} ~,
\\
\mathcal{T} &\equiv \left( 8\pi \over 3 H_0^2 \Omega_{m0} \right)^{1/2}
\simeq 75.5 \text{Gyr} ~,
\\
\mathcal{M} &\equiv {1\over 4 \pi G} \left( 8\pi \over 3 H_0^2 \Omega_{m0} \right)^{-1/4} \left( {\hbar\over m}\right)^{3/2}
\simeq 5.6 \times 10^{6} M_\odot \left( 10^{-23} \text{eV} \over m \right)^{3/2} ~,
\end{align}
and the rescaling, 
\begin{align}
\tilde{x} = {x \over \mathcal{L}} ~,
\quad
\tilde{t} = {t \over \mathcal{T}} ~,
\quad
\tilde{\Phi} = \mu \mathcal{T} \Phi ~,
\quad
\tilde{\psi} = \mathcal{T} \sqrt{4\pi G \mu} \psi ~,
\quad
\tilde{g} = {1\over \mathcal{T} 4\pi G \mu} g ~.
\end{align}
and we also have relationships, 
\begin{align}
\mathcal{T} = \mu \mathcal{L}^2 ~,
\quad
\mathcal{M} = \mathcal{T}^{-2} \mathcal{L}^{3} {1 \over 4 \pi G} ~.
\end{align}


**Then, SP eq.becomes dimensionless, and**
\begin{align}
i { \partial \over \partial \tilde{t}} \tilde{\psi}
&=
- {1 \over 2} \tilde{\nabla}^2 \tilde{\psi}
+ \tilde{g} |\tilde{\psi}|^2 \tilde{\psi}
+ \tilde{\psi} \tilde{\Phi} ~,
\\
\tilde{\nabla}^2 \tilde{\Phi} &=  |\tilde{\psi}|^2 - \tilde{n}_0  ~.
\end{align}


The rescaled quantities are given by
\begin{align}
\tilde{n} &= |\tilde{\psi}|^2 = \tilde{\rho} =  \mathcal{T}^2 4 \pi G \mu n = {\mathcal{L}^3 \over \mathcal{M}} \mu n ~,
\\
\tilde{N} &= \int \mathrm{d}^3 \tilde{\mathbf{x}} |\tilde{\psi}|^2 = {\mathcal{T}^2 4 \pi G \mu \over \mathcal{L}^3} N = \mathcal{L} 4 \pi G \mu^3 N ~,
\\
\tilde{v} &= v {\mathcal{T} \over \mathcal{L}} ~,
\\
\tilde{\lambda}_\text{dB} &= { 2 \pi \over \tilde{v}} ~,
\\
M &= \mathcal{M} \int \mathrm{d}^3\tilde{x} |\tilde{\psi}|^2 ~,
\\
\rho &= \mathcal{M} \mathcal{L}^{-3} |\tilde{\psi}|^2
\simeq 3.16 \times 10^{-9} M_\odot ~\text{pc}^{-3} |\tilde{\psi}|^2 ~,
\\
E &= \mathcal{M} \mathcal{L}^2 \mathcal{T}^{-2} E_\text{code} ~,
\\
{\partial\over\partial t} &= {1\over \mathcal{T}} {\partial\over\partial \tilde{t}} ~,
\\
\nabla &= {1\over \mathcal{L}} \tilde{\nabla}~,
\\
\tilde{g} &= {1\over \mathcal{T} 4\pi G \mu} g 
\simeq 3.5 \times 10^5 \left( {g\over 10^{-21} \text{GeV}^{-2}} \right) \left( \mu \over 10^{-23} \text{eV} \right)^{-1}
\\&\simeq - 4.4 \times 10^{-16} \left( {f_a \over 10^{10} \text{GeV}} \right)^{-2} \left( \mu \over 10^{-23} \text{eV} \right)^{-1}
\end{align}

The time step should satisfy the CFL condition to ensure certain precision,
\begin{align}
\Delta t \leq \max\left[ {\mu \over 6 \hbar} \Delta x^2, {\hbar \over \mu |V|_\text{max} } \right] ~,
\end{align}
which gives
\begin{align}
\Delta \tilde{t} \leq {1 \over 6} \Delta \tilde{x}^2
\end{align}
in most cases.



The initial condition can be chosen as a delta-function distribution $|\psi_{\mathbf{k}}|^2 = {N \over 4\pi k_0^2} \delta(|\mathbf{k}| - k_0)$ or a Gaussian distribution $|\psi_{\mathbf{k}}|^2 = {N \over \pi^{3/2} } e^{-|\mathbf{k}|^2}$, both with a random phase. Perfoming the Fourier transformation on $\psi_{\mathbf{k}} e^{i S}$, where $S$ is a random phase, $\psi(0,\mathbf{x})$ is homogeneous and isotropic in real space. Note that we have the identity $\int d^3\mathbf{x} |\psi(t, \mathbf{x})|^2 = {A \over B} \int d^3\mathbf{k} |\psi_{\mathbf{k}}(t)|^2$, where $A$, $B$ refer to Fourier convention, and we take $A=B$ here for simplicity.

The momentum and Fourier component are rescaled as
\begin{align}
\tilde{\mathbf{k}} &= {\mathbf{k} \over k_0} ~,
\\
\tilde{\psi}_{\tilde{\mathbf{k}}} &= (4 \pi G)^{1/2} k_0^{3/2} \mu^{3/2} L^{1/2} \psi_\mathbf{k} ~.
\end{align}


The rescaled initial conditions are written as
\begin{align}
|\tilde{\psi}_{\tilde{\mathbf{k}}}|^2 = { \tilde{N} \over 4\pi} \delta(|\tilde{\mathbf{k}}| - 1) ~,
\quad
|\tilde{\psi}_{\tilde{\mathbf{k}}}|^2 = \pi^{-3/2} k_0^3 \tilde{N} e^{- k_0^2 |\tilde{\mathbf{k}}|^2}
\rightarrow
\pi^{-3/2} \tilde{N} e^{- |\tilde{\mathbf{k}}|^2} ~.
\end{align}


 
According to the linear perturbation theory, the Jeans length is given by
\begin{align}
k_J \equiv (16 \pi G \rho)^{1/4} \mu^{1/2} 
= (16 \pi G n \mu^3)^{1/4}
= (4 \tilde{n})^{1/4} {1\over \mathcal{L}}
\equiv \tilde{k}_J  {1\over \mathcal{L}}~.
\end{align}

$k_J = (4 \tilde{n})^{1/4}$.


The dimensionless de Broglie wavelength is given by
\begin{align}
\tilde{\lambda}_\text{dB} = { 2 \pi \over \tilde{v}} ~,
\end{align}

## Construct initial fluctuations: Gaussian

```python
#-----------------------intial Guassian distribution--------------------------------------

resol = 128   #the number of grids on each side in the real space
length = 20   #the normalized length of each side in the real space
Ntot = 100.0 #the normalized total number of bosons: N = number density * length**3

#loop calculation
actual_num_steps = 5000 #total time step
save_num = 50 #save file once every 60 timestep
ini_psi_plot=True
ini_grav_plot=True
loop_initiate=False

#--------sponge boundary condition-------------
sponge_bc = SolitonMY.sponge_func(resol, length)

start_time = time.time()

SolitonMY.soliton_formaiton(resol, length, Ntot, 
                             ini_psi_plot, ini_grav_plot, 
                             loop_initiate, actual_num_steps, save_num,sponge_bc)

print("--- %s seconds ---" % (time.time() - start_time) )
```

```python
start_time = time.time()

# Define the meta data for the movie
FFMpegWriter = matplotlib.animation.writers['ffmpeg']
metadata = dict(title='Renormalized Nnumber Density', artist='ChaoChen',comment='soliton')
writer = FFMpegWriter(fps= 2, metadata=metadata, bitrate= 300)
    

# Initialize the movie
plt.ioff()
fig = plt.figure()

data2d = []
for ix in range(actual_num_steps):
    if (ix%save_num==0):
        file_name = "psi_#{}.npy".format(ix,'npy')
        data = np.load(os.path.join('soliton_data', file_name)) # which is psi
        temp2d = (np.real(np.abs(data)**2)).sum(2) # sum |psi|^2 along z axis
        data2d.append(temp2d)    

#----------------------------------------------2d-------------------------------------
def animate2d(i):
    plt.clf()
    plot = plt.imshow(data2d[i], cmap="jet", 
                               interpolation="nearest", origin="lower", norm=matplotlib.colors.LogNorm(1e-1,1e+2),
                               extent=[-length/2.0, length/2.0, -length/2.0, length/2.0])
    plt.colorbar()
    
    plt.contour(data2d[i], levels=[1e+2],
            extent=[- length/2.0, length/2.0, -length/2.0, length/2.0], 
            colors = "black", origin = 'lower')
    
# Update the frames for the movie
with writer.saving(fig, "Soliton_2d.mp4", 200):
    for ix in range(len(data2d)):
        animate2d(ix)
        writer.grab_frame()


print("--- %s seconds ---" % (time.time() - start_time) )



#--------------------------------------------------------------------------------------
# don't worry about the code in this cell, it is just to let you 
# display the movies you generated above in Jupyter notebook
from IPython.display import HTML

HTML("""
<div align="middle">
<video width="80%" controls> 
      <source src="Soliton_2d.mp4" type="video/mp4">
</video></div>""")
```

```python
# profile of the formed soliton

file_name = "psi_#{}.npy".format(4950,'npy')
data = np.load(os.path.join('soliton_data', file_name)) # load psi data
rho_data = (np.real(np.abs(data)**2)).sum(2) # sum along z axis
max_val = np.max( rho_data ) # max of |psi|^2
max_pos = np.unravel_index( np.argmax( rho_data ), rho_data.shape ) # max position
print("the maximum value:",rho_data[max_pos],rho_data[max_pos])
print("the maximum postion:",max_pos)

norm_data = rho_data/rho_data[max_pos] # normalize data using max

#-----------------define distance function for 2d--------------------------
grid_vec = np.linspace( - length/2.0 + length/ float(2.0*resol), 
                       length/2.0 - length/float(2.0*resol), resol)

x_2d, y_2d = np.meshgrid(grid_vec, grid_vec, sparse=False, indexing ='ij') #(resol, 1) for x
dist_2d = ( ( x_2d - x_2d[max_pos[0],0] )**2 + ( y_2d - y_2d[0,max_pos[1]] )**2)**0.5 # distance from max position.
if dist_2d[max_pos] ==0.0:
    print("distance function is well-defined")
else:
    print("distance function is incorrect")
    

#---------------------average data points----------------------------------
def ave_func(dist_2d,norm_data,interval):
    dist_x = dist_2d.reshape(resol**2)
    norm_data_y = norm_data.reshape(resol**2) # raw data;
        
    bin_edges = np.arange(np.min(dist_x),np.max(dist_x), interval) # group data points for averaging      
    bin_indices = np.digitize(dist_x, bin_edges) # group indices

    new_x = []
    new_y = []
    for kk in range(1,np.max(bin_indices)+1):
        indices = np.where(bin_indices==kk)[0] # find the corresponding indices of y data
        if len(indices)==0:
            pass
        else:
            yy = np.sum(norm_data_y[indices])/len(indices) # sum up values in a same bin
            xx = dist_x[indices[0]]
            new_y.append(yy)
            new_x.append(xx)
    return new_x, new_y

data_x, data_y = ave_func(dist_2d,norm_data,0.1)

plt.scatter(dist_2d, norm_data, color='grey',s=0.1)
plt.scatter(data_x, data_y, color='red',s=5)
plt.xlabel('r')
plt.ylabel('$|\psi|^2/|\psi_{max}|^2$')
plt.xscale('log')
plt.yscale('log')
plt.ylim(1e-3,1e+0)
plt.show()
```
